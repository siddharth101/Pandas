## Accessing rows, rows and columns
To access rows we can use iloc and loc
iloc integer location based while loc uses label based indexing. In the current dataframe, both will yield the same result

```python
### Selecting one row and all columns
df_movies.iloc[0,:] # same as df_movies.iloc[0]
```

```python
### Selecting two rows or more rows and all columns
df_movies.iloc[[0,4,8],:] # same as df_movies.iloc[[0,4,8]]
```

```python
### Selecting n rows and m columns
df_movies.iloc[[0,4],[0,1,2]]
```

```python
### This will give first 4 rows second to fourth columns
df_movies.iloc[0:4,1:4]
```

```python
df_movies.iloc[:,[4,5]] ## all rows and fifth and sixth column
```

#### Let's say we want to select first 5 rows for the columns duration and gross without worrying about the integer index for the columns duration and gross, we can use loc instead of iloc

```python
df_movies.loc[:5,['duration','gross']]
```

```python
### First four rows and all columns starting form duration and ending on gross
df_movies.loc[:5,'duration':'gross']
```

```python
### First four rows and all columns after and including gross
df_movies.loc[:5,'gross':]
```

### Conditional Statements

```python
df_movies.head()
```

### One condition: Movies directed by James Cameron

```python
df_movies[df_movies['director_name']=='James Cameron']
```

### Two conditions (AND): Movies directed by James Cameron & with imdb score above 8

```python
df_movies[(df_movies['director_name']=='James Cameron') & (df_movies['imdb_score']>8)]
```

### Two conditions (OR): Movies directed by James Cameron | Sam Mendes

```python
df_movies[(df_movies['director_name']=='James Cameron') | (df_movies['director_name']=='Sam Mendes')]
```

### Three conditions (AND,OR): Movies directed by James Cameron | Sam Mendes & imdb_score above 8

```python
c1 = df_movies['imdb_score']>8
c2 = (df_movies['director_name']=='James Cameron') | (df_movies['director_name']=='Sam Mendes')
df_movies[ (c1) & (c2)]
```
